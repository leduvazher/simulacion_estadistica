## Generador Congruencial

Implementar el generador de números pseudo aleatorios con el método congruencial, donde:



$r_{i+1}=a r_{i}mod(m) $



Siendo:



$r_{i}$: El número pseudo aleatorio en el momento $i$ que se emplea para generar $r_{i+1}$.



$r_{i+1}$: El nuevo número pseudo aleatorio generado.



$a$: Coeficiente del modelo



$m$: Valor con el que se obtendrá el módulo



$b$: Tamaño de la palabra de la computadora



$u_i$: Valor pseudo uniforme acotado entre $0<u_i<1$



$m=2^b$



$a = 2^\frac{b+1}{2}\$



$u_i = \frac{r_i}{m}\$


```python
def pseudo_uniformes(b,r_0,n):
    # Objetivo: Gemrar números pseudo aleatorios con distribución uniforme
    # Entrada:
    #      b= Tamaño de la palabra de la computadora que ejecuta el código
    #      r_0= Semilla para generar la serie de números pseudo aleatorios
    #      n: Números de valores a generar
    # Salida:
    #      lst_pseudo_uniformes: Lista de los números pseudo aleatorios con distribución uniforme
    
    lst_pseudo_uniformes = [] # inicializando la lista que tendrá el resultado final
    
    if r_0==-1:
        return lst_pseudo_uniformes
    
    m = 2**b
    a = 2**((b+1)/2)
          
    r_i = r_0 # Inicializar el valor de número pseudo aleatorio
    
    for i in range(0,n):
        r_i = (a*r_0) % m # <---- r(i+1)=a*r(i)mod m
        r_0 = r_i
        u_i = r_i / m
        
        lst_pseudo_uniformes.append(u_i)
    
    # proporcionar el resultado de los cálculos
    return lst_pseudo_uniformes
    
```

```python
# Invocar la función para generar números pseudo aleatorios con distribución uniforme
u = pseudo_uniformes(b=64, r_0=3187947384, n=30)
u
```

<h1> Método de la Transformación Inversa. Distribución Uniforme <h1>

```python
import numpy as np
```

```python
def generar_valores_uniformes(a=-1,b=-1):
    '''
    
    Objetivo: Generar valores con distrubcion uniforme utilizando la funcion
              obtenida con el Método de Transformación Inversa.
    Entradas:
              a: Valor mínmo del intervalo de la distrubión uniforme
              b: Valor máximo del intervalo de distrubución uniforme
    Salidas:
            x: Valor distribución uniforme
    
    
    '''
    
    #Validas las entradas del programa
    
    if a==-1 and b==-1:
        return 0
    elif a>b:   #else if ... elif
        return 0
    
    u = np.random.rand() #generar un valor pseudoaleatorio uniforme
    
    x = (b-a)*u + a
    
    return x

```

```python
a = 3
b = 10

for i in range(0,10):
    x=generar_valores_uniformes(a,b)
    r=round(x,0)
    print("{}: {}; {}". format(i,x,r))
```

# Distribución Bernoulli



Para simular esta distribución de probabilidad se genera **un evento aleatorio $r$** y se compara con la **probabilidad de éxito $p$**, y se toma el valor de la variable de estudio, $x$, igual a uno cuando es un éxito i.e. ($r<p$), y de cero en el caso de no serlo ($r\geq p$).



$p$: Probabilidad de éxito para el evento $x=1$



$1-p$: Probabilidad de fracaso para el evento $x=0$

```python
N = 10 # Numero de experimentos a realizar

for experimento in range(0,N):
    cnt_aguila= 0  #contador de cara=aguila
    cnt_sol   = 0 #contador de cara-sol
    
    #Evento Bernoulli
    
    if np.random.rand()<0.5:
        cnt_aguila+=1 #contador aguila = contador aguila + 1
    else:
        cnt_sol+=1
    
    p_aguila = cnt_aguila #probabilidad de que cara=aguila
    p_sol    = cnt_sol    #probabilidad de que cara = sol
    
    print("Experimento #{} Pr(aguila)={}, Pr(sol)={}".format(experimento+1,p_aguila, p_sol))
    
    #Pintar los resultados
    label = ['Aguila', 'Sol']
    data  = [p_aguila, p_sol]
    
    plt.xlabel('Clasificacion')
    plt.ylabel('Probabilidad')
    
    titulo = 'Distribucion de la Probabilidad del Experimento #' + str(experimento+1)
    plt.title(titulo)
    
    plt.bar(label, data)
    plt.show
```

# Distribución Binomial



Para simular esta distribución de probabilidad es necesario realizar N experimentos Bernoulli y sumar el resultado en cada uno. Con ésto, se acumulan los diversos experimentos cuando son exitosos. A continuación se muestra el código en Python que simula una v.a. Binomial.



$N$: Número de ensayos Bernoulli



$p$: Probabilidad de éxito para el evento $x=1$



$1-p$: Probabilidad de fracaso para el evento $x=0$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%%time

N = 10_000       #Numero de ensayos Bernoulli
cnt_exito = 0    # Contador de eventos exitosos
pr_exito  = 0.6  #Probabilidad de exito

for _ in range(N):
    u = np.random.rand() # generar un evento uniforme
    
    if u <= pr_exito:
        cnt_exito+=1
        
print("Cantidad exitos={}".format(cnt_exito))
print("Cantidad de fracasos={}".format(N - cnt_exito))
print("Probabilidad exitos={}".format(cnt_exito/N))
print("Probabilidad fracaso={}".format(1-(cnt_exito/N)))


#Graficar los datos

label = ['Exitos', 'Fracasos']
data = [cnt_exito / N , (1-cnt_exito/N)]
plt.xlabel("Clasificacion")
plt.ylabel("Probabilidad")
plt.title("Distribucion de las Probabilidades")
plt.bar(label, data)
plt.show
```
